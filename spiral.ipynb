```python
import math as m
import matplotlib.pyplot as plt
import matplotlib.animation as animation

def getNextVertex(vX, vY):
  h = m.sqrt(m.pow(vX, 2) + m.pow(vY, 2))
  d = (vX - vY/h), (vY + vX/h)
  return d
```

```python
fig, ax = plt.subplots()

ax.set_aspect('equal')
ax.axis('off')
ax.plot([0, 1], [0, 0], 'r') # draw the first side

n = 17
xOld, yOld = 1, 0

sides = []
hipotenuses = []

for i in range(1, n):
  xNew, yNew = getNextVertex(xOld, yOld)

  # sides
  side = ax.plot([xOld, xNew], [yOld, yNew], 'k')[0]

  # hiponuse
  hipotenuse = ax.plot([0, xNew], [0, yNew], 'r')[0]

  xOld, yOld = xNew, yNew

plt.show()
```
